```python
# Total sales and profit
total_sales = sal['Sales'].sum()
total_profit = sal['Profit'].sum()

# Average sales and profit per transaction
avg_sales = sal['Sales'].mean()
avg_profit = sal['Profit'].mean()

# Print results
print(f"Total Sales: {total_sales}")
print(f"Total Profit: {total_profit}")
print(f"Average Sales: {avg_sales}")
print(f"Average Profit: {avg_profit}")
```

```python
# Sales and profit by category
sales_by_category = sal.groupby('Category')['Sales'].sum()
profit_by_category = sal.groupby('Category')['Profit'].sum()

# Sales and profit by region
sales_by_region = sal.groupby('Region')['Sales'].sum()
profit_by_region = sal.groupby('Region')['Profit'].sum()

# Display results
print("Sales by Category:\n", sales_by_category)
print("Profit by Category:\n", profit_by_category)
print("Sales by Region:\n", sales_by_region)
print("Profit by Region:\n", profit_by_region)
```

```python
# Convert Date column to datetime if not already
sal['Order Date'] = pd.to_datetime(sal['Order Date'])

# Set the Date column as index
sal.set_index('Order Date', inplace=True)

# Resample by month and by year
monthly_sales = sal['Sales'].resample('M').sum()
yearly_sales = sal['Sales'].resample('Y').sum()

# Plot trends
monthly_sales.plot(title='Monthly Sales Trend')
yearly_sales.plot(title='Yearly Sales Trend')
```
